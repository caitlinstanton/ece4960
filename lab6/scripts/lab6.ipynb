# Plotting trajectories based on odometry and ground truth

You will need to **repeatedly** send points to the plottter to plot the odometry and groundtruth trajectories as the robot moves. Write your plot updation code within the function *update_plot()*, which is then subsequently called in the last line of the cell. Use the ***time.sleep()*** function to prevent updating too often. <br>

```python
def update_plot(robot):
    while True:
        start = time.perf_counter()
        pose = robot.get_pose()
        gt_pose = robot.get_gt_pose()
        robot.send_to_plot(pose[0], pose[1], ODOM)
        robot.send_to_plot(gt_pose[0], gt_pose[1], GT)
        end = time.perf_counter()
        print(end-start)
        #time.sleep(200)
    
update_plot(robot)
```
